### 소상공인코드 1 --소상공인코드 넣고 돌리기

```python
x = data_temp_2.drop('대분류숫자',axis=1) #대분류숫자를 Y값으로 설정, 나머지는 X
y = data_temp_2['대분류숫자']
```

```python
tree_model = DecisionTreeClassifier(max_depth=3) #모델링 깊이는 3
tree_model.fit(x,y)
```

```python
fn = ['money', 'channel', 'payment', 'time', 'code']  # 시각화에 쓸 이름들
cn = ['A', 'C', 'E', 'F', 'G', 'H', 'I', 'J', 'L', 'M', 'N', 'O', 'P',
       'Q', 'R', 'S']
```

```python
from sklearn import tree
from matplotlib import pyplot as plt   #트리 시각화

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=1000)

tree.plot_tree(
    tree_model, # 학습한 모형
    feature_names = fn, # 사용한 변수 이름
    class_names = cn, # 예측할 타겟 클래스 이름
    rounded=True, # 사각형 끝을 둥글게
    filled=True # 사각형 안 색깔 채우기
)

fig.savefig('imagename.png')
```
